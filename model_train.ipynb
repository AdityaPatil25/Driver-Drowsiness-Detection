```python
import tensorflow as tf
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dropout,Input,Flatten,Dense,MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
traindatahelp = ImageDataGenerator(rescale= 1./255, rotation_range=0.2,shear_range=0.2,
zoom_range=0.2,width_shift_range=0.2,height_shift_range=0.5,validation_split= 0.2)
```

```python
batchsize = 8
```

```python
train_data = traindatahelp.flow_from_directory(r'C:\\Users\\itsad\\Desktop\\ML projects\\Driver Drowziness\\Train'
,target_size =(80,80),batch_size=batchsize, class_mode='categorical', subset='training')

validation_data = traindatahelp.flow_from_directory(r'C:\\Users\\itsad\\Desktop\\ML projects\\Driver Drowziness\\Train'
,target_size =(80,80),batch_size=batchsize, class_mode='categorical', subset='validation')
```

```python
testdatahelp = ImageDataGenerator(rescale=1./255)
test_data = testdatahelp.flow_from_directory(r'C:\\Users\\itsad\\Desktop\\ML projects\\Driver Drowziness\\Test'
,target_size =(80,80),batch_size=batchsize, class_mode='categorical')
```

```python
bmodel = InceptionV3(include_top=False, weights='imagenet', input_tensor=Input(shape=(80,80,3)))
hmodel = bmodel.output
hmodel = Flatten()(hmodel)
hmodel = Dense(64, activation='relu')(hmodel)
hmodel = Dropout(0.5)(hmodel)
hmodel = Dense(2,activation='softmax')(hmodel)

model = Model(inputs=bmodel.input,outputs = hmodel)
for layer in bmodel.layers:
    layer.trainable = False
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
```

```python
checkpoint = ModelCheckpoint(r'C:\\Users\\itsad\\Desktop\\ML projects\\Driver Drowziness\\models\\best_model.keras', monitor='val_loss',
                             save_best_only=True, verbose=3)

earlystop = EarlyStopping(monitor='val_loss', patience=7, verbose=3, restore_best_weights=True)

learning_rate = ReduceLROnPlateau(monitor='val_loss', patience=3, verbose=3)

callbacks = [checkpoint,earlystop,learning_rate]
```

```python
model.compile(optimizer='Adam',loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(train_data, steps_per_epoch=train_data.samples//batchsize,validation_data= validation_data,
validation_steps= validation_data.samples//batchsize, callbacks = callbacks, epochs = 50)
```

Model Evaluation


```python
acc_tr, loss_tr = model.evaluate_generator(train_data)
print(acc_tr)
print(loss_tr)
```

```python
acc_vr, loss_vr = model.evaluate_generator(validation_data)
print(acc_vr)
print(loss_vr)
```

```python
acc_test, loss_test = model.evaluate_generator(test_data)
print(acc_test)
print(loss_test)
```
